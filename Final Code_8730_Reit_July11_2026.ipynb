```python
import pandas as pd
import yfinance as yf
import requests
import pymongo
import certifi
from IPython.display import display

# Disable scientific notation for readability
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
print("📥 STEP 1: Fetching RAW data from APIs...\n")

# ---------------------------------------------------------
# Source 1: yFinance API (Raw Market Data)
# ---------------------------------------------------------
tickers = ["SRU-UN.TO", "REI-UN.TO", "CHP-UN.TO", "FCR-UN.TO", "CAR-UN.TO"]
raw_market_data = []

for ticker in tickers:
    stock = yf.Ticker(ticker)
    raw_market_data.append(stock.info)

df_raw_market = pd.DataFrame(raw_market_data)

print("--- All Raw yFinance Data (First 5 Rows) ---")
display(df_raw_market.head())

# ---------------------------------------------------------
# Source 2: Bank of Canada Valet API (Raw Macro Data)
# ---------------------------------------------------------
boc_url = "https://www.bankofcanada.ca/valet/observations/V122530/json?recent=10"
response = requests.get(boc_url)

if response.status_code == 200:
    raw_boc_data = response.json().get("observations", [])
    df_raw_boc = pd.DataFrame(raw_boc_data)

    print("\n--- All Raw Bank of Canada Data (First 5 Rows) ---")
    display(df_raw_boc.head())
else:
    print("❌ Failed to fetch Bank of Canada data.")
```

```python
print("🧹 STEP 2: Cleaning and filtering required columns...\n")

# ---------------------------------------------------------
# Clean Market Data
# ---------------------------------------------------------
required_cols = ['symbol', 'longName', 'previousClose', 'dividendYield', 'marketCap', 'trailingPE']
available_cols = [col for col in required_cols if col in df_raw_market.columns]

df_filtered_market = df_raw_market[available_cols].copy()

# Rename columns for the database
df_filtered_market.rename(columns={
    'symbol': 'ticker',
    'longName': 'name',
    'previousClose': 'last_close_price',
    'dividendYield': 'dividend_yield',
    'marketCap': 'market_cap',
    'trailingPE': 'trailing_pe'
}, inplace=True)

# Convert Market Cap to Billions
if 'market_cap' in df_filtered_market.columns:
    df_filtered_market['market_cap_billions'] = df_filtered_market['market_cap'] / 1_000_000_000

print("--- Cleaned & Filtered Market Data ---")
display(df_filtered_market.head())

# ---------------------------------------------------------
# Clean BoC Data
# ---------------------------------------------------------
clean_boc_data = []
for index, row in df_raw_boc.iterrows():
    clean_boc_data.append({
        "date": row['d'],
        "target_rate_percentage": float(row['V122530']['v'])
    })
df_filtered_boc = pd.DataFrame(clean_boc_data)

print("\n--- Cleaned & Filtered BoC Data ---")
display(df_filtered_boc.head())
```

```python
# Data cleaning and standardization:

import pandas as pd
import pymongo
import certifi
from IPython.display import display

# 1. Connect to MongoDB
MONGO_URI = "mongodb+srv://noveshsewruttun19_db_user:1RokvH56xkrlT6gW@novcluster0.g4zrftf.mongodb.net/?appName=NovCluster0"

try:
    client = pymongo.MongoClient(MONGO_URI, tlsCAFile=certifi.where())
    db = client['smartcentres_project_final']

    print("🧹 STEP 1: Starting Data Cleaning Process...\n")

    # The collections that contain the raw CSV data
    collections_to_clean = ['company_portfolio_metrics', 'company_financial_performance']

    # 2. Define the cleaning logic
    def clean_financial_string(val):
        """Removes $, %, commas, and converts to float if it is a number."""
        if isinstance(val, str):
            # Strip out the unwanted characters
            cleaned_str = val.replace('$', '').replace('%', '').replace(',', '').strip()
            try:
                # Attempt to convert to a mathematical float
                return float(cleaned_str)
            except ValueError:
                # If it fails (e.g., text like "Net Income"), return the original text
                return val
        return val

    # 3. Process and update each collection
    for coll_name in collections_to_clean:
        collection = db[coll_name]

        # Pull data from MongoDB into a Pandas DataFrame
        data = list(collection.find())

        if not data:
            print(f"⚠️ Collection '{coll_name}' is empty. Skipping.")
            continue

        df = pd.DataFrame(data)

        # Drop the internal MongoDB '_id' so we can cleanly overwrite it later
        if '_id' in df.columns:
            df = df.drop(columns=['_id'])

        # Apply the cleaning function to every column
        for col in df.columns:
            df[col] = df[col].apply(clean_financial_string)

        print(f"--- Cleaned Data Preview for '{coll_name}' ---")
        display(df.head())

        # 4. Push the standardized data back to MongoDB
        collection.delete_many({}) # Delete the old, messy string data
        collection.insert_many(df.to_dict(orient='records')) # Insert the clean float data
        print(f"✅ Successfully cleaned and updated '{coll_name}' in MongoDB.\n")

    print("🎉 Data cleaning and standardization is complete! All numbers are now ready for analysis and charts.")

except Exception as e:
    print(f"❌ Connection or Execution Error: {e}")
```

```python
# Dividend Yield vs. Payout Ratio Analysis:

import pandas as pd
import pymongo
import certifi

# Connect to MongoDB
MONGO_URI = "mongodb+srv://noveshsewruttun19_db_user:1RokvH56xkrlT6gW@novcluster0.g4zrftf.mongodb.net/?appName=NovCluster0"
client = pymongo.MongoClient(MONGO_URI, tlsCAFile=certifi.where())
db = client['smartcentres_project_final']

print("🔍 ANALYSIS PART 1: Dividend Sustainability\n")

# 1. Fetch Financial Performance Data (for Payout Ratios)
financials_data = list(db['company_financial_performance'].find({}, {'_id': 0}))
df_fin = pd.DataFrame(financials_data)

# Extract Payout Ratio to AFFO for the full years
payout_2024 = df_fin.loc[df_fin['Financial Line Item'] == 'Payout Ratio to AFFO (%)', 'Full Year 2024'].values[0]
payout_2025 = df_fin.loc[df_fin['Financial Line Item'] == 'Payout Ratio to AFFO (%)', 'Full Year 2025'].values[0]

# 2. Fetch Market Valuation Data (for Competitor Yields)
market_data = list(db['valuation_metrics'].find({}, {'_id': 0}))
df_market = pd.DataFrame(market_data)

# Get SmartCentres Yield vs Competitors
sru_yield = df_market.loc[df_market['ticker'] == 'SRU-UN.TO', 'dividend_yield'].values[0]
competitors = df_market[df_market['ticker'] != 'SRU-UN.TO']
avg_comp_yield = competitors['dividend_yield'].mean()

# 3. Print the Analytical Output
print(f"⭐ SmartCentres Dividend Yield: {sru_yield}%")
print(f"📊 Competitor Average Yield: {avg_comp_yield:.2f}%\n")

print("--- Payout Ratio Trend (Lower is safer) ---")
print(f"2024 AFFO Payout Ratio: {payout_2024}%")
print(f"2025 AFFO Payout Ratio: {payout_2025}%")

if payout_2025 < payout_2024:
    print("\n✅ INSIGHT: The payout ratio is decreasing. This means SmartCentres is keeping more cash on hand. The 6.15% yield is highly sustainable and funded safely by their operational cash flow.")
else:
    print("\n⚠️ INSIGHT: The payout ratio is increasing. The dividend might be at risk if revenues drop.")
```

```python
# Chart 1: Stock Price Total Returns (Line Chart):

import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("📊 Generating Visualization 1: Competitor Total Returns...\n")

# Set executive styling
sns.set_theme(style="whitegrid")

# Define tickers
tickers = ["SRU-UN.TO", "REI-UN.TO", "CHP-UN.TO", "FCR-UN.TO"]
names = ["SmartCentres", "RioCan", "Choice Properties", "First Capital"]

# Fetch 1-year historical closing prices
hist_data = yf.download(tickers, period="1y")['Close']

# Normalize the prices to start at 100 (shows percentage return)
normalized_returns = (hist_data / hist_data.iloc[0]) * 100

# Create the plot
plt.figure(figsize=(12, 6))
colors = ['#b8874f', '#5c7c93', '#888888', '#cccccc'] # SmartCentres highlighted

for i, ticker in enumerate(tickers):
    if ticker == "SRU-UN.TO":
        plt.plot(normalized_returns.index, normalized_returns[ticker], label=names[i], color=colors[i], linewidth=3)
    else:
        plt.plot(normalized_returns.index, normalized_returns[ticker], label=names[i], color=colors[i], linewidth=1.5, alpha=0.7)

plt.title('1-Year Total Return Comparison (Normalized to 100)', fontsize=16, fontweight='bold', pad=15)
plt.ylabel('Normalized Value', fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.legend(title="REIT Peers", loc="upper left")
plt.axhline(y=100, color='black', linestyle='--', linewidth=1, alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Chart 2: Dual-Axis Chart (Macroeconomic Impact):

import requests
from datetime import datetime, timedelta

print("📊 Generating Visualization 2: Macroeconomic Impact (Dual-Axis)...\n")

# 1. Fetch 1-Year BoC Interest Rate Data
one_year_ago = (datetime.now() - timedelta(days=365)).strftime('%Y-%m-%d')
boc_url = f"https://www.bankofcanada.ca/valet/observations/V122530/json?start_date={one_year_ago}"
boc_response = requests.get(boc_url).json()

boc_dates = [datetime.strptime(obs['d'], '%Y-%m-%d') for obs in boc_response['observations']]
boc_rates = [float(obs['V122530']['v']) for obs in boc_response['observations']]

# 2. Fetch 1-Year SmartCentres Stock Data
sru_stock = yf.Ticker("SRU-UN.TO").history(period="1y")

# 3. Create the Dual-Axis Plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Stock Price on primary Y-axis (Left)
color1 = '#b8874f'
ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('SmartCentres Stock Price (CAD)', color=color1, fontsize=12, fontweight='bold')
ax1.plot(sru_stock.index, sru_stock['Close'], color=color1, linewidth=2.5, label="SRU Price")
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(False)

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()

# Plot BoC Rate on secondary Y-axis (Right)
color2 = '#5c7c93'
ax2.set_ylabel('Bank of Canada Policy Rate (%)', color=color2, fontsize=12, fontweight='bold')
ax2.plot(boc_dates, boc_rates, color=color2, linewidth=2.5, linestyle='--', label="BoC Rate")
ax2.tick_params(axis='y', labelcolor=color2)
ax2.grid(axis='y', linestyle='--', alpha=0.3)

plt.title('Macroeconomic Impact: SmartCentres Price vs. BoC Interest Rates', fontsize=16, fontweight='bold', pad=15)
fig.tight_layout()
plt.show()
```
